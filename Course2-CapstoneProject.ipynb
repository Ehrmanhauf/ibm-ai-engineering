# PART B

B. Normalize the data (5 marks) 

Repeat Part A but use a normalized version of the data. Recall that one way to normalize the data is by subtracting the mean from the individual predictors and dividing by the standard deviation.

How does the mean of the mean squared errors compare to that from Step A?

```python
X = concrete_data.drop('Strength', axis = 1)
y = concrete_data['Strength']

mse_list = []
for i in range(50):
    print(f"Starting the {i+1}. run.")
    
    # We perform the train-test split here.
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

    # We perform the normalization here.
    scaler = StandardScaler()
    # We will fit the scaler on the training data and use the learned scaling parameters on the test set to transform it as well.
    scaler.fit(X_train)

    # We normalize the training and test set using learned parameters from the train set.
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    
    n_cols = X.shape[1]
    
    model = Sequential()
    
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    model.fit(X_train, y_train, epochs = 50)
    
    predictions = model.predict(X_test)
    
    mse = mean_squared_error(y_test, predictions)
    print(f"The MSE of the model is {mse} on {i+1}. run.")
    mse_list.append(mse)
```

Reporting the error metrics for the case with normalized data (Part B))

```python
mean_mse = np.mean(mse_list)
std_mse = np.std(mse_list)

print(f"The mean of the mean squared errors is {mean_mse}.")
print(f"The standard deviation of the mean squared errors is {std_mse}.")
```

The mean of the mean squared errors was *198.08* and the standard deviation was *153.51* in part A, where the data was not normalized.

In part B, we see that the mean is up to *353.85* and the standard deviation is down to *76.95*. While the mean error greatly increased, there is a decrease in the standard deviation.

The decrease in the standard deviation is expected since we performed normalization on the data and we hopefully expect to eliminate the impact of outliers in some features of the data. On the other hand, we the mean error's skyrocketed, and the reason for that may be an underfitting problem. One solution can be to increase the number of epochs.

C. Increate the number of epochs (5 marks)

Repeat Part B but use 100 epochs this time for training.

How does the mean of the mean squared errors compare to that from Step B?

```python
X = concrete_data.drop('Strength', axis = 1)
y = concrete_data['Strength']

mse_list = []
for i in range(50):
    print(f"Starting the {i+1}. run.")
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

    # We perform the normalization here.
    scaler = StandardScaler()
    # We will fit the scaler on the training data and use the learned scaling parameters on the test set to transform it as well.
    scaler.fit(X_train)

    # We normalize the training and test set using learned parameters from the train set.
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    
    n_cols = X.shape[1]
    
    model = Sequential()
    
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    model.fit(X_train, y_train, epochs = 100)
    
    predictions = model.predict(X_test)
    
    mse = mean_squared_error(y_test, predictions)
    print(f"The MSE of the model is {mse} on {i+1}. run.")
    mse_list.append(mse)
```

As we expected, both mean and standard deviation of the mean squared errors lowered significantly when we increased the epochs. The model learned better, so the performance improved.

Repeat part B but use a neural network with the following instead:

```python
X = concrete_data.drop('Strength', axis = 1)
y = concrete_data['Strength']

mse_list = []
for i in range(50):
    print(f"Starting the {i+1}. run.")
    
    # We perform the train-test split here.
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

    # We perform the normalization here.
    scaler = StandardScaler()
    # We will fit the scaler on the training data and use the learned scaling parameters on the test set to transform it as well.
    scaler.fit(X_train)

    # We normalize the training and test set using learned parameters from the train set.
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    
    n_cols = X.shape[1]
    
    model = Sequential()
    
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols,)))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(1))
    
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    model.fit(X_train, y_train, epochs = 50)
    
    predictions = model.predict(X_test)
    
    mse = mean_squared_error(y_test, predictions)
    print(f"The MSE of the model is {mse} on {i+1}. run.")
    mse_list.append(mse)
```
